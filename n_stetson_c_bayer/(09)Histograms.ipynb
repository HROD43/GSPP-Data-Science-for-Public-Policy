```python
import pandas as pd
import matplotlib.pyplot as plt, numpy as np
%matplotlib inline
```

```python
df = pd.read_csv('Data/rents_with_commutes.csv')
```

```python
print(df['taz_driving'].describe())
```

```python


plt.hist(df['rent'], bins=25, color='indianred')
plt.tick_params(axis='both', 
                left='off', 
                top='off', 
                right='off', 
                bottom='off', 
                labelleft='off', 
                labeltop='off', 
                labelright='off')

plt.xlabel('Rent')
plt.title("Distribution of Rents in the Bay Area")

```

```python

plt.hist(df['taz_driving'].dropna(), bins=25, color='cadetblue')
plt.tick_params(axis='both', 
                left='off', 
                top='off', 
                right='off', 
                bottom='off', 
                labelleft='off', 
                labeltop='off', 
                labelright='off')
plt.set_xlim(0,100)
plt.xlabel('Avg Driving Commute Time')
plt.title("Distribution of Commutes in the Bay Area")

```

```python
plt.hist(df['taz_transit'].dropna(), bins=25, color='olivedrab')
plt.tick_params(axis='both', 
                left='off', 
                top='off', 
                right='off', 
                bottom='off', 
                labelleft='off', 
                labeltop='off', 
                labelright='off')
plt.set_xlim([0,100])
plt.xlabel('Avg Transit Commute Time')
plt.title("Distribution of Commutes in the Bay Area")
```

```python
plt.hist(df['taz_transit'].dropna(), bins=25, color='indianred', alpha = 0.6)
plt.hist(df['taz_driving'].dropna(), bins=25, color='cadetblue', alpha = 0.6)
plt.title("Gaussian/Uniform Histogram")
plt.xlabel("Average Commute Time")
plt.ylabel("Probability")
plt.legend()
plt.show()
```

```python
plt.figure(1, figsize=(8,8), )

ax = plt.subplot(211)
ax.set_title("Driving")
ax.set_xlabel('Avg Driving Commute, in Minutes')
#ax.set_xlim(0,100)
ax.tick_params(axis='both', 
                left='off', 
                top='off', 
                right='off', 
                bottom='off', 
                labelleft='off', 
                labeltop='off', 
                labelright='off')
plt.hist(df['taz_driving'].dropna(), bins=25, color='cadetblue', linewidth = 0.3)

ax = plt.subplot(212)
ax.set_title("Transit")
ax.set_xlabel('Avg Transit Commute, in Minutes')
#ax.set_xlim(0,100)
ax.tick_params(axis='both', 
                left='off', 
                top='off', 
                right='off', 
                bottom='off', 
                labelleft='off', 
                labeltop='off', 
                labelright='off')
plt.hist(df['taz_transit'].dropna(), bins=25, color='olivedrab', linewidth=0.3)



# Normalize command to set axis??

plt.subplots_adjust(wspace=.5, hspace=.5)
plt.show()
```
